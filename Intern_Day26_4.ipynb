```python
X_train,Y_train=smote.fit_resample(X_train,Y_train)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import accuracy_score, classification_report

# Convert to numpy if pandas
X_train = np.array(X_train)
X_test = np.array(X_test)
Y_train = np.array(Y_train)
Y_test = np.array(Y_test)

# Reshape for CNN: (samples, channels, features)
X_train = X_train.reshape(X_train.shape[0], 1, X_train.shape[1])
X_test = X_test.reshape(X_test.shape[0], 1, X_test.shape[1])

# Dataset
class TabularDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.long)

    def __len__(self):
        return len(self.y)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

train_dataset = TabularDataset(X_train, Y_train)
test_dataset = TabularDataset(X_test, Y_test)

train_loader = DataLoader(train_dataset, batch_size=256, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=256, shuffle=False)

# Optimized CNN Model
class CNN1D(nn.Module):
    def __init__(self, input_size, num_classes):
        super(CNN1D, self).__init__()

        self.conv1 = nn.Conv1d(1, 32, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm1d(32)

        self.conv2 = nn.Conv1d(32, 64, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm1d(64)

        self.pool = nn.MaxPool1d(2)

        # Corrected calculation: pooling divides the final sequence length by 2
        conv_output_size = (input_size // 2) * 64

        self.fc1 = nn.Linear(conv_output_size, 128)
        self.dropout = nn.Dropout(0.3)
        self.fc2 = nn.Linear(128, num_classes)

    def forward(self, x):
        # Apply layers cleanly
        x = torch.relu(self.bn1(self.conv1(x)))
        x = torch.relu(self.bn2(self.conv2(x)))
        x = self.pool(x)  # Pooled after conv2 to align with conv_output_size

        x = x.view(x.size(0), -1)

        x = torch.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)

        return x

# Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

num_classes = len(np.unique(Y_train))
input_size = X_train.shape[2]

model = CNN1D(input_size, num_classes).to(device)

criterion = nn.CrossEntropyLoss()

# FIX: Switched to standard learning rates and momentum for faster minimization
optimizer = optim.Adam(model.parameters(), lr=0.001, betas=(0.9, 0.999))

# Training
from sklearn.metrics import f1_score
import matplotlib.pyplot as plt

epochs = 100

train_losses = []
f1_scores = []

for epoch in range(epochs):

    model.train()
    running_loss = 0

    for inputs, labels in train_loader:

        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()

        outputs = model(inputs)
        loss = criterion(outputs, labels)

        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    epoch_loss = running_loss / len(train_loader)
    train_losses.append(epoch_loss)

    # -----------------------
    # Calculate F1 Score
    # -----------------------
    model.eval()

    epoch_preds = []
    epoch_labels = []

    with torch.no_grad():

        for inputs, labels in test_loader:

            inputs = inputs.to(device)

            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)

            epoch_preds.extend(preds.cpu().numpy())
            epoch_labels.extend(labels.numpy())

    epoch_f1 = f1_score(
        epoch_labels,
        epoch_preds,
        average='weighted'
    )

    f1_scores.append(epoch_f1)

    print(
        f"Epoch [{epoch+1}/{epochs}] "
        f"Loss: {epoch_loss:.4f} "
        f"F1: {epoch_f1:.4f}"
    )
```

```python
plt.figure(figsize=(8,6))

plt.plot(
    train_losses,
    f1_scores,
    marker='o'
)

plt.xlabel("Loss")
plt.ylabel("F1 Score")
plt.title("Loss vs F1 Score")

plt.grid(True)

plt.show()
```
