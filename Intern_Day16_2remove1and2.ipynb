```python
from imblearn.over_sampling import SMOTE;
smote=SMOTE();
```

```python
Xoversampled,Yoversampled=smote.fit_resample(X_train,Y_train)
```

```python
from xgboost import XGBClassifier


xgb=XGBClassifier()


estimators=[('xgb',xgb)]


from sklearn.pipeline import Pipeline


pipe=Pipeline(steps=estimators)


pipe
```

```python
from skopt import BayesSearchCV


from skopt.space import Real,Categorical,Integer


search_space = {
    'max_depth': Integer(3, 10),
    'learning_rate': Real(0.01, 0.3, prior='log-uniform'),
    'n_estimators': Integer(400, 500),
    'subsample': Real(0.6, 1.0),
    'colsample_bytree': Real(0.6, 1.0),
    'reg_alpha': Real(0.0, 5.0),
    'reg_lambda': Real(0.1, 5.0),
    'gamma': Real(0.0, 5.0)
}

opt=BayesSearchCV(pipe,search_space,n_iter=10,cv=3,scoring='roc_auc',random_state=42)


opt = BayesSearchCV(
    xgb,
    search_spaces=search_space,
    n_iter=10,
    cv=3
)
```

```python
opt.fit(Xoversampled,Yoversampled)
```

```python
Ypred=opt.predict(Xoversampled)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
accuracy_score(Yoversampled,Ypred)
```

```python
classification_report(Yoversampled,Ypred)
```

```python
confusion_matrix(Yoversampled,Ypred)
```
