```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import accuracy_score, classification_report

# Convert to numpy if pandas
X_train = np.array(X_train)
X_test = np.array(X_test)
Y_train = np.array(Y_train)
Y_test = np.array(Y_test)

# Reshape for CNN: (samples, channels, features)
X_train = X_train.reshape(X_train.shape[0], 1, X_train.shape[1])
X_test = X_test.reshape(X_test.shape[0], 1, X_test.shape[1])

# Dataset
class TabularDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.long)

    def __len__(self):
        return len(self.y)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]

train_dataset = TabularDataset(X_train, Y_train)
test_dataset = TabularDataset(X_test, Y_test)

# Increased batch size to 256 for faster execution steps
train_loader = DataLoader(train_dataset, batch_size=256, shuffle=True, pin_memory=True)
test_loader = DataLoader(test_dataset, batch_size=256, shuffle=False, pin_memory=True)

# CNN 1D Model
class CNN1D(nn.Module):
    def __init__(self, input_size, num_classes):
        super(CNN1D, self).__init__()

        self.conv1 = nn.Conv1d(1, 32, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm1d(32)

        self.conv2 = nn.Conv1d(32, 64, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm1d(64)

        self.pool = nn.MaxPool1d(2)

        # FIX: Added adaptive pooling so it scales seamlessly across any number of tabular features
        self.adaptive_pool = nn.AdaptiveAvgPool1d(16) 
        conv_output_size = 64 * 16 

        self.fc1 = nn.Linear(conv_output_size, 128)
        self.dropout = nn.Dropout(0.3)
        self.fc2 = nn.Linear(128, num_classes)

    def forward(self, x):
        x = self.pool(torch.relu(self.bn1(self.conv1(x))))
        x = torch.relu(self.bn2(self.conv2(x)))
        x = self.adaptive_pool(x) # Ensures predictable shape before flattening

        x = x.view(x.size(0), -1)

        x = torch.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)

        return x

# Device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

num_classes = len(np.unique(Y_train))
input_size = X_train.shape[2]

model = CNN1D(input_size, num_classes).to(device)

criterion = nn.CrossEntropyLoss()  # Adjust weights if needed
# Increased learning rate to 0.001 for quicker initial convergence
optimizer = optim.Adam(model.parameters(), lr=0.001, betas=(0.9, 0.999)) 

# Dynamic Scheduler to aggressively reduce loss towards 0.1
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=3)
# Scaler for Automatic Mixed Precision (AMP) acceleration
scaler = torch.amp.GradScaler() 

# Training
epochs = 100

for epoch in range(epochs):
    model.train()
    running_loss = 0.0

    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device, non_blocking=True), labels.to(device, non_blocking=True)

        # set_to_none=True clears gradients faster than zero_grad()
        optimizer.zero_grad(set_to_none=True) 

        # Forward pass using Mixed Precision (AMP) for speed boost
        with torch.amp.autocast(device_type=device.type):
            outputs = model(inputs)
            loss = criterion(outputs, labels)

        # Scaled backward pass
        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        running_loss += loss.item()

    epoch_loss = running_loss / len(train_loader)
    print(f"Epoch [{epoch+1}/{epochs}] Loss: {epoch_loss:.4f}")

    # Pass loss to scheduler to drop learning rate when training slows down
    scheduler.step(epoch_loss)

    # Automated exit criteria once loss is near or below 0.1
    if epoch_loss <= 0.105:
        print(f"Target loss (~0.1) achieved! Stopping training early at Epoch {epoch+1}.")
        break

# Evaluation
model.eval()

all_preds = []
all_labels = []

with torch.no_grad():
    for inputs, labels in test_loader:
        inputs = inputs.to(device)

        # Match training precision style during inference
        with torch.amp.autocast(device_type=device.type):
            outputs = model(inputs)
            
        _, preds = torch.max(outputs, 1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.numpy())

accuracy = accuracy_score(all_labels, all_preds)

print(f"\nTest Accuracy: {accuracy:.4f}")
print("\nClassification Report:")
print(classification_report(all_labels, all_preds))
```
